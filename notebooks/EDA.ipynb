### What is the main analysis/modelling goal?

For this project, the main goal will be **model or predict average house prices**. With this in mind, organise features into feature blocks.

---

- **A: House Type Features**

| Price Features | Index Features |
| :------------- | :------------- |
| ```DetachedPrice``` | ```DetachedIndex``` |
| ```SemiDetachedPrice``` | ```SemiDetachedIndex``` |
| ```TerracedPrice``` | ```TerracedIndex``` |
| ```FlatPrice``` | ```FlatIndex``` |

- Important features for defined goal.

---

- **B: Buyer Type Features**

| Price | Index |
| :------------- | :------------- |
| ```CashPrice``` | ```CashIndex``` |
| ```MortgagePrice``` | ```MortgageIndex``` |
| ```FTBPrice``` | ```FTBIndex``` |
| ```FOOPrice``` | ```FOOIndex``` |

- Drop these to simplify.
- Interesting for policy and social analysis.
- Separate out for possible future investigations for buyer behaviours.

---

- **C: Property Age Features**

| Price Features | Index Features |
| :------------- | :------------- |
| ```NewPrice``` | ```NewIndex``` |
| ```OldPrice``` | ```OldIndex``` |

- Same logic as above: separate for possible future work, but not immediately relevant.

---

- **D: General Price and Index**

| Feature | Description |
| :------------- | :------------- |
| ```AveragePrice``` | Mean sale price |
| ```Index``` | Price index relative to base period |
| ```AveragePriceSA``` | Seasonally adjusted price |
| ```IndexSA``` | Seasonally Adjusted Index |

- Seasonal adjustments have high correlation with unadjusted versions.
- Seasonal adjustments also have limited data coverage.
- Drop seasonally adjusted columns.

---

- **E: Sales Volume**
- Useful for context, or modelling demand.


````python
df.drop(columns=['AveragePriceSA', 'IndexSA'], inplace=True)  # Seasonally adjusted columns
df.drop(columns=['CashPrice', 'MortgagePrice', 'FTBPrice', 'FOOPrice', 'CashIndex', 'MortgageIndex', 'FTBIndex', 'FOOIndex'], inplace=True)  # Buyer Type Features
df.drop(columns=['NewPrice', 'OldPrice', 'NewIndex', 'OldIndex'], inplace=True)  # Property Age Features
df.drop(columns=['DetachedIndex', 'SemiDetachedIndex', 'TerracedIndex', 'FlatIndex'], inplace=True)  # Indices
````

````python
df['CashShare'] = df['CashSalesVolume'] / df['SalesVolume']
df['MortgageShare'] = df['MortgageSalesVolume'] / df['SalesVolume']
df['NewShare'] = df['NewSalesVolume'] / df['SalesVolume']
df['OldShare'] = df['OldSalesVolume'] / df['SalesVolume']
````

````python
df.drop(columns=['CashSalesVolume', 'MortgageSalesVolume', 'NewSalesVolume', 'OldSalesVolume'], inplace=True)
````

# Data Cleaning & Feature Engineering

## Handle Missing Values

````python
df.isnull().sum().sort_values(ascending=False)
````

### Interpolate missing numerical data

````python
df[['1m%Change', '12m%Change', 'SalesVolume']] = df[['1m%Change', '12m%Change', 'SalesVolume']].interpolate(method='linear')
````

## Target Encoding

Encode each region with the average target value (```AveragePrice```).

````python
region_avg_price = df.groupby('RegionName')['AveragePrice'].mean()
df['RegionEncoded'] = df['RegionName'].map(region_avg_price)
````

## Extract datetime features

### New Features

````python
df = df.sort_values(by=['RegionName', 'Date'])
df['PriceChange'] = df.groupby('RegionName')['AveragePrice'].pct_change()
````

````python
df['DetachedToAvgRatio'] = df['DetachedPrice'] / df['AveragePrice']
df['SemiDetachedToAvgRatio'] = df['SemiDetachedPrice'] / df['AveragePrice']
df['TerracedToAvgRatio'] = df['TerracedPrice'] / df['AveragePrice']
df['FlatToAvgRatio'] = df['FlatPrice'] / df['AveragePrice']
````

````python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter
df.drop(columns=['Date'], inplace=True)
````

````python
df.drop(columns=['RegionName', 'AreaCode'], inplace=True)
````

````python
df.to_csv('../data/processed.csv.gz', index=False, compression='gzip')
````
